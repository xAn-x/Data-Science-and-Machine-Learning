```python
# In matpotlib we specifically want pylot->
import matplotlib.pyplot as plt
```

```python
# There are 2 types of graph
    # -> Points->use points to show the graph
    # -> connected->connected the scattered point

# To plot a graph we need x & y cordn so that we can scatter it
x=[1,2,3]
y=[2,4,6]
```

```python
# Points-Graph->

# First tell pyplot how points are scattered in cartesian plane-> scatter(x_cords,y_cords)
plt.scatter(x,y)

# Now to display graph -> show()
plt.show()
```

```python
# Connected-Graph->

# First tell pyplot what points needed to be plot-> plot(x_cord,y_cord)
plt.plot(x,y)
plt.show()
```

```python
# U can do both scattering and ploting perception simountaneously
plt.scatter(x,y)
plt.plot(x,y)

plt.show()
```

```python
# We can plot more than one graph also->

# Graph-1
x1=[i for i in range(10)]
y1=[i**2 for i in range(10)]

# Graph-2
x2=[i for i in range(10)]
y2=[-(i**2) for i in range(10)]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()
```

```python
x=[i for i in range(5)]
y=[i for i in range(5)]

plt.plot(x,y,'ro--')
plt.show()

# -- -> dotted line
# - -> dash line
# o/p -> points
# ^ -> Tiangle
# -. -> combn of . and -   etc
# * -> star marker 
```

```python
# Question->plot graph of x^3 in domain [-10,10)

import numpy as np

x_dim=np.arange(-10,10,0.5)
y_dim=x_dim**3

plt.plot(x_dim,y_dim,".-")

plt.show()
```

```python
# Plotting Trigo graphs
x=np.arange(-4*np.pi,4*np.pi,0.1)
y=np.sin(x)

plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.title('SIN(X)')
plt.plot(x,y,color='green')
plt.show()
```

```python
# Can plot no of distinct graph using subplot(no of rows,no of cols,position)

# Sin Graph
plt.subplot(2,1,1)

x=np.arange(-4*np.pi,4*np.pi,0.1)
y=np.sin(x)

plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.title('SIN(X)')
b=plt.plot(x,y,color='green')

 
plt.show()

# Cos Graph

plt.subplot(2,1,2)

x=np.arange(-4*np.pi,4*np.pi,0.1)
y=np.cos(x)

plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.title('COS(X)')
plt.plot(x,y,color='red')

a=plt.show(2)
```
