```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(-10,10,0.5)
y=x**3

plt.plot(x,y,".-")
plt.show()
```

```python
# Adding cordinates axes to our graph

# To plot axis use axhline(x-axis) and axvline(y-axis)

# axhline(x_cord,color)/axvline(y_cord,color)

plt.plot(x,y,".-")
plt.axhline(0,color="grey")
plt.axvline(0,color="grey")

plt.show()
```

```python
# We can provide custom color and marker and can set line width also
plt.plot(x,y,color='black',marker='.',linewidth=0.5)

plt.axhline(0,color="grey")
plt.axvline(0,color="grey")

plt.show()
```

```python
# Now We can provide labels to axis also -> like which axis corresponds to what

# xlabel("x_axis_name")/ylabel("y_axis_name")
plt.plot(x,y,color='black',marker='.',linewidth=0.5)

plt.xlabel("X")
plt.ylabel("X^3")

plt.axhline(0,color="grey")
plt.axvline(0,color="grey")

plt.show()
```

```python
# Now id i have more than 1 graph it is very usefull to know what plot representing what graph for that
    # First label all your graph
    # Then use a function legend() to show them
    
x=np.arange(-5,5,0.5)

y1=x**3
y2=x**2

plt.plot(x,y1,color='black',marker='.',linewidth=0.5,label="x^3")
# plt.subplot(2,2,2)
plt.plot(x,y2,color='orange',marker='.',linewidth=0.5,label="x^2")

plt.axhline(0)
plt.axvline(0)

# grid()->Provide grids in the cartesian-plain
plt.grid()

plt.legend()
plt.show()
```

```python
# can add text to yor graph also

# text(x_cor,y_cord,"Text",fontSize='--')
x=np.arange(-5,5,0.5)
y=x**3

plt.plot(x,y,color='orange')
plt.text(3,5,'Text Added')

# use to provide title to your char
plt.title('x^3 chart')

plt.show()
```

----- BUBBLE-CHART -----

```python
# Let say we need to represent data showing average-salary of a country on following years with count of working population
year=[2012,2013,2014,2015,2016,2017]
salary=(np.array([12,13,14,10,17,18]))*10
population=[100,120,140,100,150,165]
```

```python
# Using Scatter Graph
plt.scatter(year,salary,label="salary")
plt.scatter(year,population,label="population")

plt.legend()
plt.show()
```

```python
# Using plot graph
plt.plot(year,salary,"--",label="salary")
plt.plot(year,population,"-",label="population")

plt.show()
```

```python
# Using Bubble graph->

# Bubble graph are shatter graph with size of each point thats'it

# To represent size use argume 's'->
plt.scatter(year,salary,label="salary",s=population,color='orange')
plt.show()
```

```python
# can change color of each bubble just provide an array with size=number of points and pass color=arr in argument
colors=np.arange(len(year))

plt.scatter(year,salary,label="salary",s=population,c=colors)
plt.show()

# can change edge-colors of marker using edgecolor=''
```

```python
# To make them transparent or opaque use argument->alpha
plt.scatter(year,salary,label="salary",s=population,color='green',alpha=0.5)
plt.show()
```

------ PIE-CHARTS -------

```python
# Draw a char to show how many childrens like what kind of Fruit

partitions=[10,2,8,15,5,1]
plt.pie(partitions)

plt.show()
```

```python
# Let say i want to axis it equally
plt.pie(partitions)

plt.axis('equal')
plt.show()
```

```python
# Let Say i want to add labels to it to know what represent what and custom colors
fruit_names=['Apple','Kiwi','Mango','strawBerry','WaterMelon','Grapes']
colors=['red','green','yellow','#FF69B4','lime','purple']
plt.pie(partitions,labels=fruit_names,colors=colors)

plt.title('Fav-Fruits')
plt.show()
```

```python
# If I want to know what percentage of students like a fruit

# use->autopct=func/or/"upto what decimal we need to show data"
plt.pie(partitions,labels=fruit_names,colors=colors,autopct="%.2f%%")
plt.show()
```

```python
#If we want to explode a piece out just use explode=[how much to explode]

explode=[0.2,0,0,0.1,0,0]

plt.pie(partitions,labels=fruit_names,colors=colors,autopct="%.2f%%",explode=explode)
plt.show()
```

```python
# The graph is counterclock-wise in nature but u cna change it just pass argument->counterclock=False

# We can also provide startangle which tell from where to start first slice
explode=[0.2,0,0,0.1,0,0]

plt.pie(partitions,labels=fruit_names,colors=colors,autopct="%.2f%%",explode=explode,counterclock=False,startangle=120)
plt.axis('equal')
plt.show()
```

------- HISTOGRAM --------

```python
data=np.array([1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14])
```

```python
# edgecolor use to specify the edge of 1 ine
plt.hist(data,edgecolor='black')
plt.show()
```

```python
# We can divide the histogram into x-ticks to chk for what x-val our bar is there
xticks=np.arange(1,23)
plt.hist(data,edgecolor='black')

plt.xticks(xticks)
plt.show()
```

```python
# As we can see some bars/bins are clubbed together to form 1 bar->as by-default it produce 10 bars

# size of bin=(maxNo-minNo)/no_of_bins

# We can divide the histogram into x-ticks to chk for what x-val our bar is there
xticks=np.arange(1,23)

# always divide chart inot maxNo-minNo of bins
plt.hist(data,edgecolor='black',bins=21)

plt.xticks(xticks)
plt.show()

# Note in a bin/tower [st,end) this order is followed
```

```python
# We can pass our custom bins also to get corressponfing histo
plt.hist(data,bins=[1,3,6,10,15,20],edgecolor='white')
plt.show()
```

```python
# We can set axis len also
# axis([x_st,x_end,y_st,y_end])
plt.hist(data,bins=[1,3,6,10,15,20],edgecolor='white')
plt.axis([0,25,0,15])
plt.show()
```
