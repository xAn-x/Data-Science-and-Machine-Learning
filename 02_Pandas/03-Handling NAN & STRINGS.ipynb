--------- Handling NAN-dataValues ---------

```python
import pandas as pd
import numpy as np
```

```python
with open("Iris_dataset.csv") as Iris_dataset:
    iris=pd.read_csv(Iris_dataset)

iris.head()
```

```python
# We don't have any NAN values in our Iris_dataset so making some values NAN
iris.iloc[1:5,2:5]=np.nan
iris.iloc[13:19,1:3]=np.nan



iris.head()
```

```python
iris.describe()
```

```python
# Now since data-set contain some NAN values we have to handle that

temp=iris.copy()

# Method-1=>Droping it

# dropna()->This function return a new data-fram with removing rows where nan is +nt 
iris.dropna(inplace=True)
print(iris.head())

print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")

# U can see some data-changes 
iris.describe()
```

```python
iris=temp.copy()
iris.head(7)
```

```python
# Method-2=>Flling Nan with some sort of values

# It's always good Idea to fill NAN with mean of the col/type ehich it affecting but u can use any arbitary value also

# fillna()->return  a new data-set by filling all values of the column with passed values
iris.SepalWidthCm.fillna(iris.SepalWidthCm.mean(),inplace=True)
iris.head(7)
```

--------- Handaling STRINGS in data --------

Sometimes We have to deal with string in our data but working with string is kind a complicate like 
how to find mean , how to perform any Mathematical calculation etc...

And this string info can be really crucila so need to handel and etract info from it also

```python
iris=temp.copy()

# Let's add a column 'Gender' so we can handel it carefully
iris['Gender']='Female'
iris.iloc[1:7,6]='Male'
iris.iloc[13:27,6]='Male'

iris.head(10)
```

```python
# Now Let's say we are marking 0 for male and 1 for female

# We can apply a function over a column which will do all this stuff for us and add a new column corresponding to that
def get_sexType(gender):
    if gender=='Male':
        return 0
    else:
        return 1
    
iris['sex']=iris.Gender.apply(get_sexType)

# We can simply delete gender-column as we have our sex_type column
del iris['Gender']

iris.head(10)
```

```python
# Now we can apply oprn on string data also
iris.sex.mean()
```
