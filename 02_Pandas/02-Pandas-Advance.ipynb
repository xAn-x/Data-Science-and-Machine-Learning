```python
import pandas as pd
```

```python
with open("Iris_dataset.csv") as Iris_dataset:
    iris=pd.read_csv(Iris_dataset)
iris.head()
```

```python
# drop(row-no)->return a new data frame with removal of passed row-no
iris.drop(0,inplace=True)
iris.head()
```

```python
# if we try droping the same row again it will throw an error as it work according to index and not position
iris.drop(0,inplace=True)
iris

# so drop() ->remove the row but not reset the indexing
```

```python
# To know all indexes u have u can use ->index
iris.drop(3,inplace=True)
print(iris.index) # so u will not have to 0 & 3 as u dropped it
```

```python
# U can chk what index is at pos 'k' -> df.index[k]
print(iris.index[3])
```

```python
iris.head()
```

```python
# Now using index we can drop row at a particular pos also
iris.drop(iris.index[3],inplace=True)
iris.head()
```

```python
# can drop multiple rows also
iris.drop(iris.index[[0,2,4]],inplace=True)
iris
```

```python
print(iris.Species=='Iris-setosa')

print("\n\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n\n")

print(iris[iris.Species=='Iris-setosa'][:10])

print("\n\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n\n")
iris[iris.Species=='Iris-setosa'].describe()
```

```python
# NOTE-> iloc work on position and not on indexes....Will explain in Advance Pandas NOTEBOOK

print(iris.head())

print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

print(iris.iloc[0]) #it will give data of index viz on very top as use position for accessing data
```

```python
# loc is similar to iloc but it work according to index and not the position of data in data-frame
print(iris.loc[2])
```

```python
iris.tail()
```

```python
# We can use loc method to add a row at very end of your data-frame
iris.loc[0]=[1,2,3.3,5,4,"Iris-setosa"]
iris.tail()
```

```python
temp=iris.copy()
temp.head()
```

```python
# reset_index()-> Use to rest index and provide increasing order. Return a new data-frame if need to do it inplace make inplace=True
temp.reset_index()
```

```python
# Note how reset_index() store previous index in index col sepreately if we don't want that drop it
temp.reset_index(drop=True,inplace=True)
print(temp.index)
temp.head()
```

```python
temp.head()
```

```python
# If we need to remove a column we can use drop() but provide axis=1(stands for column)
temp.drop('SepalLengthCm',axis=1,inplace=True)
temp.head()
```

```python
del temp['PetalWidthCm']
temp.head()
```

Converting dataFrame to some file like CSV

```python
import numpy as np
dfrm=pd.DataFrame(np.arange(20).reshape(4,5),columns=np.array(["Column-"+str(i) for i in range(5)]),index=np.array([i for i in range(4)]))

dfrm
```

```python
# Change our dataFrame to csv 
dfrm.to_csv('my Data-Frame.csv')

# can change to xl,csv etc..
```

DataFrame to Array

```python
# Select all required values using iloc -> use .values to get ther value in a array format
array=dfrm.iloc[:].values
print(array)

print("\n\n")

# Select all required values using loc -> use .values to get ther value in a array format
array2=dfrm.loc[:].values
print(array2)
```

```python
# Can convert a part to array also
array3=dfrm.iloc[2:,1:4]
print(array3)
```
