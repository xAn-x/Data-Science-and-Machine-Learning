```python
import pandas as pd
```

```python
with open("CoinTracking.csv") as cryptoFile:
    df=pd.read_csv(cryptoFile)
# head(n=5)-> Print First n rows of data-frame
df.head(5)

# tail(n=5)-> Print last n rows of data-frame  
```

```python
# df["col-name"] / df.col_name -> use to access column with passed name
print(df['Type'])

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

# shape -> tell structure of data-frame
print(df.shape)

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

# dtypes->tell datatype of all headers data->
print(df.dtypes)

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

# columns->gives all the column in our data-frame
print(df.columns)

# can change columns also->
orgnl_headers=df.columns

df.columns=['a','b','c','d','e','f','g','h']
print()
print(df.columns)

df.columns=orgnl_headers



print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

print(f"Type of data-frame-> {type(df['Buy'])}")

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

# can print selective columns also->
print(df[{'Type','Exchange','Date'}])
```

```python
# Slicing a data-frame is same as slicing a list
print(df['Type'][:3])
```

```python
# Mathematical Operations->

# max()/min()
print("MAx: ",df['Buy'][:5].max(),"  Min: ",df['Buy'][:5].min())

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

# mean()
print('Mean: ',df['Buy'][:5].mean())

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

# std()->standard deviation
print("standard deviation: ",df["Buy"].std())

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

# describe() -> provide all neccesary information about the data-frame
df.describe()
```

```python
# Can Use Boolean Indexing to access data also->
print("Date where price>=1:\n",df['Date'][df.Buy>=1])
```

```python
# index->return the index column of df
df.index

# Can change it also to some other column using set_index

# set_index('column-name')->return a new d.f & make passed col as indexing col
temp=df.set_index('Date') 
print(temp.head(3))

print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

print(df[:3])

print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

# To do it inplace set argument inplace=True
df.set_index('Date',inplace=True)
print(df[:3])

print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

# reset_index()->return a new data-set with original index
df.reset_index(inplace=True)
print(df[:3])
```

```python
# Some more important function->

# isnull() -> tells weather a data is null or not
print(df.isnull())

print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

# To know how many total null entries we have
print(df.isnull().sum())
```

```python
# iloc -> use to acess a paricular slice of data-frame Purely integer-location based indexing for selection by position.

# .iloc[] is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array
df.iloc[0:6,3:7]
```
