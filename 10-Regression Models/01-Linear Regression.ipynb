```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
X,y=load_iris(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=12)
```

`Training Our Linear Regression Model`

```python
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
plt.subplot(1,2,1)
plt.plot(X_test,y_test)
plt.title("Real-Outputs")

plt.subplot(1,2,2)
plt.plot(X_test, y_pred)
plt.title("Predicted-Outputs")
```

```python
# Used for cross-validation and the scoring criteria used is mean_squared_error
mse_score=cross_val_score(model,X,y,scoring="neg_mean_squared_error",cv=5)
mean_mse=mse_score.mean()

print(f"Mean of MSE:{mean_mse}")

# Mean of MSE should be as close to 0
```
