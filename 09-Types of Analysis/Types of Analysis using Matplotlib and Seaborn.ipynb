```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

`Loading Dataset`

```python
iris=load_iris()
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df["target"]=iris.target
iris_df.head()
```

```python
def getSpecie(num):
    return iris["target_names"][num]
iris_df["species"]=iris_df["target"].map(lambda num: getSpecie(num))
iris_df.head()
```

`Dividing Dataset according to Specie`


```python
iris_setosa=iris_df[iris_df["species"]=="setosa"]
iris_versicolor = iris_df[iris_df["species"] == "versicolor"]
iris_virginica = iris_df[iris_df["species"] == "virginica"]
```

# Univariate Analysis

```python
plt.plot(iris_setosa["sepal length (cm)"],
         np.zeros_like(iris_setosa["petal length (cm)"]),"o")
plt.plot(iris_versicolor["sepal length (cm)"],
         np.zeros_like(iris_versicolor["petal length (cm)"]),"*")
plt.plot(iris_virginica["sepal length (cm)"],
         np.zeros_like(iris_virginica["petal length (cm)"]),".")

plt.xlabel("Sepal-Length (cm)")
```

This certainly not give a good picture about the flowers pattern

`Bivariate Analysis`

```python
plt.plot(iris_setosa["sepal length (cm)"],
         iris_setosa["petal length (cm)"],"o")
plt.plot(iris_versicolor["sepal length (cm)"],
         iris_versicolor["petal length (cm)"], "*")
plt.plot(iris_virginica["sepal length (cm)"],
         iris_virginica["petal length (cm)"], ".")

plt.xlabel("Sepal-Length (cm)")
plt.ylabel("Petal-Length (cm)")
```

This certainly not give a good picture about the flowers pattern

`Multivariate Analysis`
